```python
import geopandas as gpd
import pandas as pd
import rasterio
from rasterio.features import rasterize
from shapely.geometry import LineString, MultiLineString

# File paths
rivers_fp = '/Users/jpnousu/WBT_data/shapes/clipped_MTK-virtavesi_19-01-23_virtavesialue.shp'
streams_fp = '/Users/jpnousu/WBT_data/shapes/clipped_MTK-virtavesi_19-01-23_virtavesikapea.shp'
ref_raster_fp = '/Users/jpnousu/WBT_data/korkeusmalli_16m.tif'
out_raster_fp = '/Users/jpnousu/WBT_data/rasters/rivers_streams_bool.tif'
out_vector_fp = '/Users/jpnousu/WBT_data/shapes/merged_rivers_streams.shp'

# Load shapefiles
rivers = gpd.read_file(rivers_fp)
streams = gpd.read_file(streams_fp)

# Load reference raster for CRS, transform, shape
with rasterio.open(ref_raster_fp) as ref:
    ref_crs = ref.crs
    ref_transform = ref.transform
    ref_shape = (ref.height, ref.width)
    profile = ref.profile

# Reproject to match DEM
rivers = rivers.to_crs(ref_crs)
streams = streams.to_crs(ref_crs)

# Convert river polygons to boundary lines
rivers['geometry'] = rivers.geometry.boundary
rivers['source'] = 'river'
streams['source'] = 'stream'

# Merge
merged = pd.concat([rivers, streams], ignore_index=True)
merged = merged[['source', 'geometry']]

# Drop Z-dimension
def to_2d(geom):
    if geom.has_z:
        if isinstance(geom, LineString):
            return LineString([(x, y) for x, y, z in geom.coords])
        elif isinstance(geom, MultiLineString):
            return MultiLineString([
                LineString([(x, y) for x, y, z in line.coords]) for line in geom.geoms
            ])
    return geom

merged['geometry'] = merged.geometry.apply(to_2d)

# Explode MultiLineStrings into individual LineStrings
merged = merged.explode(ignore_index=True)

# Confirm geometries are LineString
assert all(merged.geom_type == 'LineString'), "Some geometries are not LineString after processing"

# Save shapefile
merged.to_file(out_vector_fp)

# Rasterize to boolean raster (1 for river/stream, 0 otherwise)
raster = rasterize(
    [(geom, 1) for geom in merged.geometry],
    out_shape=ref_shape,
    transform=ref_transform,
    fill=0,
    dtype='uint8'
)

# Write raster
profile.update(dtype='uint8', count=1, compress='lzw', nodata=0)
with rasterio.open(out_raster_fp, 'w', **profile) as dst:
    dst.write(raster, 1)
```

```python
import geopandas as gpd
import pandas as pd
import rasterio
from rasterio.features import rasterize
from shapely.geometry import LineString, MultiLineString
import shapely

# File paths
rivers_fp = '/Users/jpnousu/WBT_data/shapes/clipped_MTK-virtavesi_19-01-23_virtavesialue.shp'
streams_fp = '/Users/jpnousu/WBT_data/shapes/clipped_MTK-virtavesi_19-01-23_virtavesikapea.shp'
ref_raster_fp = '/Users/jpnousu/WBT_data/korkeusmalli_16m.tif'
out_raster_fp = '/Users/jpnousu/WBT_data/rasters/rivers_streams_bool.tif'
out_vector_fp = '/Users/jpnousu/WBT_data/shapes/merged_rivers_streams_2D.shp'

# Load shapefiles
rivers = gpd.read_file(rivers_fp)
streams = gpd.read_file(streams_fp)

# Load reference raster for CRS, transform, shape
with rasterio.open(ref_raster_fp) as ref:
    ref_crs = ref.crs
    ref_transform = ref.transform
    ref_shape = (ref.height, ref.width)
    profile = ref.profile

# Reproject rivers and streams to DEM CRS
rivers = rivers.to_crs(ref_crs)
streams = streams.to_crs(ref_crs)

# Convert river polygons to boundaries (optional)
rivers['geometry'] = rivers.geometry.boundary
rivers['source'] = 'river'
streams['source'] = 'stream'

# Merge rivers and streams
merged = pd.concat([rivers, streams], ignore_index=True)
merged = merged[['source', 'geometry']]

# 🔥 Drop Z-values to make sure we have 2D lines
def to_2d(geom):
    if hasattr(geom, "has_z") and geom.has_z:
        if isinstance(geom, LineString):
            return LineString([(x, y) for x, y, _ in geom.coords])
        elif isinstance(geom, MultiLineString):
            return MultiLineString([
                LineString([(x, y) for x, y, _ in line.coords]) for line in geom.geoms
            ])
    return geom

merged['geometry'] = merged.geometry.apply(to_2d)

# 🌊 Explode MultiLineStrings to individual LineStrings
merged = merged.explode(ignore_index=True)

# 💡 Ensure all geometries are 2D LineStrings
merged = merged[merged.geom_type == 'LineString'].copy()

# ❗ Save cleaned 2D shapefile
merged.to_file(out_vector_fp)

# ✅ Rasterize for visual/analysis comparison
raster = rasterize(
    [(geom, 1) for geom in merged.geometry],
    out_shape=ref_shape,
    transform=ref_transform,
    fill=0,
    dtype='uint8'
)

# Save boolean raster
profile.update(dtype='uint8', count=1, compress='lzw', nodata=0)
with rasterio.open(out_raster_fp, 'w', **profile) as dst:
    dst.write(raster, 1)
```
