```python
import os
import shutil
import rasterio

files_to_convert = ['/Users/jpnousu/WBT_data/rasters/dem_16/twi_dinf.tif',
                   '/Users/jpnousu/WBT_data/rasters/dem_16/twi_d8.tif',
                   '/Users/jpnousu/WBT_data/rasters/dem_16/flowacc_dinf.tif',
                   '/Users/jpnousu/WBT_data/rasters/dem_16/flowacc_d8.tif',
                   '/Users/jpnousu/WBT_data/rasters/dem_16/slope.tif',
                   '/Users/jpnousu/WBT_data/rasters/dem_16/dem_raw.tif',
                   '/Users/jpnousu/WBT_data/rasters/dem_16/dem_no_deps.tif',
                   '/Users/jpnousu/WBT_data/rasters/dem_16/catchment_mask.tif'] # THESE SHOULD BE FIRST CONVETED TO .ASC

converted_files = []
for tif_file in files_to_convert:
    asc_file = os.path.splitext(tif_file)[0] + '.asc'  # replace .tif with .asc
    with rasterio.open(tif_file) as src:
        meta = src.meta.copy()
        data = src.read(1)

    meta.update({
        "driver": "AAIGrid",  # ASCII Grid
        "dtype": data.dtype
    })

    with rasterio.open(asc_file, "w", **meta) as dst:
        dst.write(data, 1)

    converted_files.append(asc_file)

files = [
    '/Users/jpnousu/WBT_data/rasters/rasanen/LAI_grass.asc',
    '/Users/jpnousu/WBT_data/rasters/rasanen/LAI_shrub.asc',
    '/Users/jpnousu/WBT_data/rasters/vmi_16/LAI_decid.asc',
    '/Users/jpnousu/WBT_data/rasters/vmi_16/LAI_pine.asc',
    '/Users/jpnousu/WBT_data/rasters/vmi_16/LAI_spruce.asc',
    '/Users/jpnousu/WBT_data/rasters/vmi_16/canopy_fraction.asc',
    '/Users/jpnousu/WBT_data/rasters/vmi_16/canopy_height.asc',
    '/Users/jpnousu/WBT_data/rasters/vmi_16/sitefertility_mnfi.asc',
    '/Users/jpnousu/WBT_data/rasters/vmi_16/maintype_mnfi.asc',
    '/Users/jpnousu/WBT_data/rasters/soil_16/bottom_soil_gtk.asc',
    '/Users/jpnousu/WBT_data/rasters/soil_16/top_soil_gtk.asc',
    '/Users/jpnousu/WBT_data/rasters/dem_16/twi_dinf.asc',
    '/Users/jpnousu/WBT_data/rasters/dem_16/twi_d8.asc',
    '/Users/jpnousu/WBT_data/rasters/dem_16/flowacc_dinf.asc',
    '/Users/jpnousu/WBT_data/rasters/dem_16/flowacc_d8.asc',
    '/Users/jpnousu/WBT_data/rasters/dem_16/slope.asc',
    '/Users/jpnousu/WBT_data/rasters/dem_16/dem_raw.asc',
    '/Users/jpnousu/WBT_data/rasters/dem_16/dem_no_deps.asc',
    '/Users/jpnousu/WBT_data/rasters/dem_16/catchment_mask.asc']

out_dir = r'/Users/jpnousu/SpaFHy_RUNS/pallas/gis/pallasjarvi_16'
os.makedirs(out_dir, exist_ok=True)  # make sure output directory exists

for f in files:
    filename = os.path.basename(f)
    out_path = os.path.join(out_dir, filename)
    shutil.copy(f, out_path)
```
